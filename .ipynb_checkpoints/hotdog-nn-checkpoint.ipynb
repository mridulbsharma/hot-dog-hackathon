# Hot Dog or Not Hot Dog Classifier
## A Comprehensive Approach Using Multiple CNN Architectures


```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16, ResNet50, InceptionV3, EfficientNetB0, MobileNetV2, DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

## 1. Data Preparation
### Set up data generators

```python
# Set up data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    'data/hotdog-nothotdog/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
test_generator = test_datagen.flow_from_directory(
    'data/hotdog-nothotdog/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)
```

```python
print("Data generators set up successfully.")
print(f"Training samples: {train_generator.samples}")
print(f"Test samples: {test_generator.samples}")
```

## 2. Model Creation

```python
def create_model(base_model, model_name):
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    
    for layer in base_model.layers:
        layer.trainable = False
    
    model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
    return model, model_name
```

```python
%%time
# Create models
models = [
    create_model(VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'VGG16'),
    create_model(ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'ResNet50'),
    create_model(InceptionV3(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'InceptionV3'),
    create_model(EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'EfficientNetB0'),
    create_model(MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'MobileNetV2'),
    create_model(DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3)), 'DenseNet121')
]
```

```python
print("Models created successfully:")
for _, model_name in models:
    print(f"- {model_name}")
```

## 3. Model Training

```python
histories = []

for model, model_name in models:
    print(f"\nTraining {model_name}...")
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=10,
        validation_data=test_generator,
        validation_steps=test_generator.samples // test_generator.batch_size
    )
    histories.append((history, model_name))
    model.save(f'{model_name}_hotdog_classifier.h5')
    print(f"{model_name} training complete and model saved.")
```

## 4. Visualization of Training Progress

```python
# Function to create subplots for each model
def plot_training_progress(histories, metric):
    fig, axs = plt.subplots(3, 2, figsize=(20, 30))
    fig.suptitle(f'Model {metric.capitalize()}', fontsize=16, y=1.02)
    
    for i, (history, model_name) in enumerate(histories):
        row = i // 2
        col = i % 2
        ax = axs[row, col]
        
        ax.plot(history.history[metric], label='Training')
        ax.plot(history.history[f'val_{metric}'], label='Validation')
        ax.set_title(model_name)
        ax.set_ylabel(metric.capitalize())
        ax.set_xlabel('Epoch')
        ax.legend(loc='best')
    
    # Remove any unused subplots
    for i in range(len(histories), 6):
        fig.delaxes(axs[i // 2, i % 2])
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95)  # Adjust the top margin
    plt.show()
```

```python
# Plot accuracy
plot_training_progress(histories, 'accuracy')
```

```python
# Plot loss
plot_training_progress(histories, 'loss')
```

```python
print("Training progress visualization complete.")
```

## 5. Model Evaluation

```python
evaluation_results = []

for model, model_name in models:
    print(f"\nEvaluating {model_name}...")
    test_loss, test_acc = model.evaluate(test_generator)
    print(f'{model_name} - Test accuracy: {test_acc:.4f}')
    evaluation_results.append((model_name, test_acc))

    # Generate predictions
    predictions = model.predict(test_generator)
    predicted_classes = (predictions > 0.5).astype(int).reshape(-1)
    true_classes = test_generator.classes

    # Confusion Matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Classification Report
    print(f"\nClassification Report - {model_name}")
    print(classification_report(true_classes, predicted_classes, target_names=['Not Hot Dog', 'Hot Dog']))
```

## 6. Model Comparison

```python
comparison_df = pd.DataFrame(evaluation_results, columns=['Model', 'Accuracy'])
comparison_df = comparison_df.sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Accuracy', y='Model', data=comparison_df)
plt.title('Model Accuracy Comparison')
plt.xlabel('Accuracy')
plt.show()

print("\nModel Accuracy Ranking:")
print(comparison_df.to_string(index=False))
```

## 7. Best Model Selection

```python
best_model_name = comparison_df.iloc[0]['Model']
best_model = tf.keras.models.load_model(f'{best_model_name}_hotdog_classifier.h5')

print(f"The best performing model is: {best_model_name}")
print(f"Best model accuracy: {comparison_df.iloc[0]['Accuracy']:.4f}")
```

```python
# Save the best model
best_model.save('best_hotdog_classifier.h5')
print("Best model saved as 'best_hotdog_classifier.h5'")

print("\nModel training, evaluation, and selection complete.")
```
