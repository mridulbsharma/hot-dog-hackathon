# Hot Dog or Not Hot Dog Classifier
## A Comprehensive Approach Using Multiple CNN Architectures

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3, DenseNet121, MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import os
import random
from typing import Tuple, List
```

```python
# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

### 1. Data Preparation and Exploration

```python
# Define data paths
train_dir = 'data/hotdog-nothotdog/train'
test_dir = 'data/hotdog-nothotdog/test'

def count_images(directory: str) -> int:
    """Count the number of images in a directory and its subdirectories."""
    return sum([len(files) for r, d, files in os.walk(directory) 
                if any(file.lower().endswith(('.png', '.jpg', '.jpeg')) for file in files)])
```

```python
# Count images in train and test sets
train_hotdog_count = count_images(os.path.join(train_dir, 'hotdog'))
train_nothotdog_count = count_images(os.path.join(train_dir, 'nothotdog'))
test_hotdog_count = count_images(os.path.join(test_dir, 'hotdog'))
test_nothotdog_count = count_images(os.path.join(test_dir, 'nothotdog'))

print(f"Training set - Hot dogs: {train_hotdog_count}, Not hot dogs: {train_nothotdog_count}")
print(f"Test set - Hot dogs: {test_hotdog_count}, Not hot dogs: {test_nothotdog_count}")
```

```python
# Set up data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Use 20% of training data for validation
)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Create generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(299, 299),  # InceptionV3 input size
    batch_size=32,
    class_mode='binary',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(299, 299),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(299, 299),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

print("Data generators set up successfully.")
```

```python
# After creating the generators
print("Class indices:", train_generator.class_indices)
```

### 2. Model Creation

```python
def create_model(base_model, model_name: str) -> Tuple[Model, str]:
    """Create a model with a given base and add custom top layers."""
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    
    for layer in base_model.layers:
        layer.trainable = False
    
    model.compile(optimizer=Adam(learning_rate=0.0001), 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    return model, model_name
```

```python
# Create models
models = [
    create_model(InceptionV3(weights='imagenet', include_top=False, input_shape=(299, 299, 3)), 'InceptionV3'),
    create_model(DenseNet121(weights='imagenet', include_top=False, input_shape=(299, 299, 3)), 'DenseNet121'),
    create_model(MobileNetV2(weights='imagenet', include_top=False, input_shape=(299, 299, 3)), 'MobileNetV2')
]

print("Models created successfully:")
for _, model_name in models:
    print(f"- {model_name}")
```

### 3. Model Training

```python
def train_model(model: Model, model_name: str, epochs: int = 30) -> Tuple[Model, str, dict]:
    """Train a given model and return the trained model and its history."""
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
    
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=epochs,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size,
        callbacks=[early_stopping, reduce_lr]
    )
    
    model.save(f'{model_name}_hotdog_classifier.h5')
    print(f"{model_name} training complete and model saved.")
    
    return model, model_name, history.history
```

```python
histories = []

for model, model_name in models:
    trained_model, name, history = train_model(model, model_name)
    histories.append((history, name))
    models[models.index((model, model_name))] = (trained_model, name)
```

### 4. Visualization of Training Progress

```python
def plot_training_progress(histories: List[Tuple[dict, str]], metrics: List[str]):
    """
    Plot training progress for multiple models with larger subplots.
    
    Args:
    histories: List of tuples containing training history and model name
    metrics: List of metrics to plot (e.g., ['accuracy', 'loss'])
    """
    num_models = len(histories)
    num_metrics = len(metrics)
    
    # Increase figure size
    fig, axes = plt.subplots(num_metrics, num_models, figsize=(8*num_models, 6*num_metrics))
    fig.suptitle('Model Training Progress', fontsize=20, y=1.02)
    
    for i, metric in enumerate(metrics):
        for j, (history, model_name) in enumerate(histories):
            ax = axes[i, j] if num_metrics > 1 else axes[j]
            
            ax.plot(history[metric], label='Training', linewidth=2)
            ax.plot(history[f'val_{metric}'], label='Validation', linewidth=2)
            ax.set_title(f'{model_name} - {metric.capitalize()}', fontsize=16)
            ax.set_ylabel(metric.capitalize(), fontsize=14)
            ax.set_xlabel('Epoch', fontsize=14)
            ax.tick_params(axis='both', which='major', labelsize=12)
            ax.legend(loc='best', fontsize=12)
            
            # Add grid for better readability
            ax.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.95, hspace=0.3, wspace=0.3)  # Adjust space between subplots
    plt.show()
```

```python
# Plot accuracy and loss
plot_training_progress(histories, ['accuracy', 'loss'])

print("Training progress visualization complete.")
```

### 5. Model Evaluation

```python
def evaluate_model(model: Model, model_name: str) -> Tuple[float, np.ndarray, np.ndarray]:
    """Evaluate a model and return its accuracy, predictions, and true labels."""
    test_loss, test_acc = model.evaluate(test_generator)
    print(f'{model_name} - Test accuracy: {test_acc:.4f}')
    
    predictions = model.predict(test_generator)
    predicted_classes = (predictions > 0.5).astype(int).reshape(-1)
    true_classes = test_generator.classes
    
    return test_acc, predicted_classes, true_classes
```

```python
evaluation_results = []

for model, model_name in models:
    print(f"\nEvaluating {model_name}...")
    test_acc, predicted_classes, true_classes = evaluate_model(model, model_name)
    evaluation_results.append((model_name, test_acc, predicted_classes, true_classes))

    # Confusion Matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Classification Report
    print(f"\nClassification Report - {model_name}")
    print(classification_report(true_classes, predicted_classes, target_names=['Not Hot Dog', 'Hot Dog']))
```

### 6. Model Comparison

```python
comparison_df = pd.DataFrame([(name, acc) for name, acc, _, _ in evaluation_results], 
                             columns=['Model', 'Accuracy'])
comparison_df = comparison_df.sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Accuracy', y='Model', data=comparison_df)
plt.title('Model Accuracy Comparison')
plt.xlabel('Accuracy')
plt.show()
```

```python
print("\nModel Accuracy Ranking:")
print(comparison_df.to_string(index=False))
```

```python
# 7. Ensemble Method

print("\n7. Ensemble Method")
```

```python
def ensemble_predict(models: List[Model], x) -> np.ndarray:
    """Make predictions using an ensemble of models."""
    predictions = [model.predict(x) for model, _ in models]
    return np.mean(predictions, axis=0)
```

```python
ensemble_predictions = ensemble_predict(models, test_generator)
ensemble_classes = (ensemble_predictions > 0.5).astype(int).reshape(-1)
ensemble_accuracy = np.mean(ensemble_classes == test_generator.classes)
```

```python
ensemble_accuracy
```

```python
print(f'Ensemble model - Test accuracy: {ensemble_accuracy:.4f}')
```

```python
# Confusion Matrix for Ensemble
cm = confusion_matrix(test_generator.classes, ensemble_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Ensemble Model')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
# Classification Report for Ensemble
print("\nClassification Report - Ensemble Model")
print(classification_report(test_generator.classes, ensemble_classes, target_names=['Not Hot Dog', 'Hot Dog']))
```

### 8. Fine-tuning

```python
def fine_tune_model(model: Model, model_name: str, epochs: int = 10) -> Tuple[Model, dict]:
    """Fine-tune a model by unfreezing some layers and training with a low learning rate."""
    for layer in model.layers[-20:]:
        layer.trainable = True
    
    model.compile(optimizer=Adam(learning_rate=1e-5), 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    
    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-7)
    
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size,
        epochs=epochs,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size,
        callbacks=[early_stopping, reduce_lr]
    )
    
    model.save(f'{model_name}_fine_tuned.h5')
    print(f"{model_name} fine-tuning complete and model saved.")
    
    return model, history.history
```

```python
# Fine-tune the best model (assuming it's the first in the sorted list)
best_model, best_model_name = models[0]
fine_tuned_model, fine_tuned_history = fine_tune_model(best_model, best_model_name)
```

```python
# Evaluate fine-tuned model
fine_tuned_acc, fine_tuned_pred, true_classes = evaluate_model(fine_tuned_model, f"{best_model_name} (Fine-tuned)")
```

### 9. Final Model Comparison

```python
final_comparison = comparison_df.copy()
final_comparison = pd.concat([
    final_comparison,
    pd.DataFrame({
        'Model': [f"{best_model_name} (Fine-tuned)", 'Ensemble'],
        'Accuracy': [fine_tuned_acc, ensemble_accuracy]
    })
], ignore_index=True)

final_comparison = final_comparison.sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Accuracy', y='Model', data=final_comparison)
plt.title('Final Model Accuracy Comparison', fontsize=16)
plt.xlabel('Accuracy', fontsize=14)
plt.ylabel('Model', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.tight_layout()
plt.show()
```

```python
print("\nFinal Model Accuracy Ranking:")
print(final_comparison.to_string(index=False))
```
